```python
import matplotlib.pyplot as plt
import pandas as pd 
plt.style.use('ggplot')
```

```python
# CPU BENCHMARK TIMES
names = ['p0', 'p50', 'p90', 'p95', 'p99', 'p100']
base_cpu = [467.968, 753.664, 1037.312, 1118.208, 1523.712, 15417.344]
node_exporter_cpu = [481.280, 788.480, 1103.872, 1265.664, 2244.608, 20381.696] # 5%
zabbix_exporter_cpu = [486.912, 763.904, 1050.624, 1144.832, 1523.712, 20840.448] # 1%
ebpf_exporter_cpu = [471.552, 745.472, 1035.264, 1107.968, 1435.648, 21364.736] #0%

d = {'names': names, 'base': base_cpu, 'node_exporter': node_exporter_cpu, 'zabbix': zabbix_exporter_cpu, 'ebpf_exporter': ebpf_exporter_cpu}
df = pd.DataFrame(data=d)

ax = df.plot.bar(x='names', figsize=(20,10), rot=0, fontsize=20)
ax.set_ylabel("us/op", fontsize=20)
ax.set_xlabel("percentyl", fontsize=20)
ax.legend(fontsize=15)
plt.savefig("cpu_benchmark.png")
```

```python
# CPU BENCHMARK TIMES WITHOUT MAX
ax = df[df['names'] != 'p100'].plot.bar(x='names', figsize=(20,10), rot=0, fontsize=20)
ax.set_ylabel("us/op", fontsize=20)
ax.set_xlabel("percentyl", fontsize=20)
ax.legend(fontsize=15)
plt.savefig("cpu_benchmark_without_max.png")
```
